## Plotly - Menambahkan Dimensi ke Matrix Scatterplot di Plotly
Berikut ini cara menambahkan dimensi ke matrix scatterplot di plotly.

```python
#memasukan modul yang digunakan
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import plotly.offline as pyo
```

```python
#memasukan data
orders = pd.read_excel('sales.xls')
orders.head()
```

```python
#memilih variabel apa saja yang akan digunakan
orders_data = orders.loc[:, ['Sales', 'Profit', 'Discount', 'Order Priority']]
orders_data.head(2)
```

```python
#membuat figure
figure = ff.create_scatterplotmatrix(orders_data, diag = 'box', index = 'Order Priority', colormap = 'Portland', colormap_type = 'cat')
```

```python
#melihat hasilnya
pyo.plot(figure, filename = 'Add Variabel to Scatter Plot Matrix.html')
```
