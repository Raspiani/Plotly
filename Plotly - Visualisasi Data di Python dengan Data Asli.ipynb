## Plotly - Visualisasi Data di Python dengan Data Asli

Berikut ini cara membuat visualisasi data di python menggunakan data asli. 

```python
#memasukan modul yang digunakan
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
#membuka data
data = pd.read_excel('sales.xls')
data.head()
```

```python
#membuat scatter chart 
data = [go.Scatter(x = data.index, y = data.Sales, mode = 'markers')]
```

```python
#membuat layout
layout = go.Layout(
                    title = 'Sales vs Profit Chart', 
                    xaxis = dict(title = 'Sales in K'),
                    yaxis = dict(title = 'Profit in K'),
                    hovermode = 'closest'
                  )
```

```python
#membuat frame
figure = go.Figure(data = data, layout = layout)
```

```python
#Melihat hasilnya
pyo.plot(figure)
```
