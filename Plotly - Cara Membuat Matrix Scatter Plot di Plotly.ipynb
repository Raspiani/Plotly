## Plotly - Cara Membuat Matrix Scatter Plot di Plotly
Berikut ini cara membuat matrix scatter plot di plotly.

```python
#memasukan modul yang digunakan
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import plotly.offline as pyo
```

```python
#memasukan data
orders = pd.read_excel('sales.xls')
orders.head()
```

```python
#memilih variabel apa saja yang akan digunakan
orders_data = orders.loc[:, ['Sales', 'Profit', 'Discount']]
orders_data.head(2)
```

```python
#membuat kolom baru yang diberi nama index
orders_data['index'] = np.arange(1, len(orders_data)+1)

#melihat hasilnya
orders_data.head()
```

```python
#membuat figure
figure = ff.create_scatterplotmatrix(orders_data, diag = 'box', index = 'index', colormap = 'Portland', colormap_type = 'cat')
```

```python
#melihat hasilnya
pyo.plot(figure, filename = 'Scatter Plot Matrix.html')
```
