## Plotly - Line Chart di Plotly

Berikut ini cara membuat line chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#melihat info data
fifa.info()
```

```python
#membuat pivot tabel
fifa_pivot = pd.pivot_table(fifa, values = 'HomeGoals', index = 'Year')
fifa_pivot.head()
```

```python
#mendefinisikan data di plotly
data = [go.Scatter(
        x = fifa_pivot.index,
        y = fifa_pivot.HomeGoals,
        mode = 'lines',
        name = 'Average Home Goals by Year'
)]
```

```python
#mendefinisikan layout
layout = go.Layout(title = 'Average Home Goals by Year')
```

```python
#mendefinisikan frame
figure = go.Figure(data = data, layout = layout)
```

```python
#melihat hasil
pyo.plot(figure, filename = 'average_of_fifa.html')
```
