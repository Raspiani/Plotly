## Plotly - Pengenalan Plotly Visualisasi Data dan Membuat Plotly Scatter Chart

Berikut ini pengenalan plotly visualisasi data dan cara membuat plotly scatter chart. 

```python
#memasukan modul yang digunakan
import plotly.offline as pyo
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
#membuka data
data = pd.read_excel('sales.xls')
data.head()
```

```python
#Cara I
#membuat plotly scatter chart 
pyo.plot([
    go.Scatter(
    x = data.index,
    y = data.Sales, 
    mode = 'markers'
    )
])

#hasilnya seperti gambar di bawah ini
```

```python
#cara II
data = go.Scatter(x = data.index, y = data.Sales, mode = 'markers')

#Membuat plotly scatter plot
pyo.plot([data])
```
