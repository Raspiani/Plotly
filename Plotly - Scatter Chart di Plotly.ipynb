## Plotly - Scatter Chart di Plotly

Berikut ini cara membuat scatter chart di plotly.

Referensi : https://dash.plot.ly/?_ga=2.221052315.2054414018.1576406660-399155170.1567829979

```python
#memasukan modul yang digunakan
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data 
data = pd.read_excel('sales.xls')
data.head()
```

```python
#menampilkan data kolom sales dan profit
data1 = data[['Sales', 'Profit']]
data1.head()
```

```python
#mendefine data di plotly
data = [go.Scatter(
        x = data1.Sales,
        y = data1.Profit,
        mode = 'markers',
        marker = dict(size = 14, 
                      color = 'rgb(120,150,100)', 
                      symbol = 'square',
                      line = dict(width = 2)
)    
)]
```

```python
#membuat layout
layout = go.Layout(
        title = 'Sales dan Profit Chart',
        xaxis = dict(title = 'Sales di X'),
        yaxis = dict(title = 'Profit di Y'),
        hovermode = 'closest'
)
```

```python
#membuat frame
figure = go.Figure(data = data, layout = layout)
```

```python
#menampilkan hasil
pyo.plot(figure)
```
