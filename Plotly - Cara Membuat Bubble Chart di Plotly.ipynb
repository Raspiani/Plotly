## Plotly - Cara Membuat Bubble Chart di Plotly
Berikut ini cara membuat bubble chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
orders = pd.read_excel('sales.xls')
orders.head()
```

```python
#mendefinisikan data di plotly
data = [go.Scatter(
    x = orders['Sales'],
    y = orders['Profit'],
    text = orders['Customer Name'],
    mode = 'markers',
    marker = dict(size = 100 * orders['Discount'], color = '#E9967A')
)]
```

```python
#membuat layout
layout = go.Layout(
    title = 'Sales vd Profit',
    xaxis = dict(title = 'Sales'),
    yaxis = dict(title = 'Profit'),
    hovermode = 'closest'
)
```

```python
#mendefinisikan figure
figure = go.Figure(data = data, layout = layout)
```

```python
#melihat hasilnya
pyo.plot(figure, filename = 'Bubble Chart di Plotly.html')
```
