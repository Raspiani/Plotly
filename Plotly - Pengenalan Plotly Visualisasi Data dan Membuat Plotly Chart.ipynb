## Plotly - Pengenalan Plotly Visualisasi Data dan Membuat Plotly Chart

Berikut ini pengenalan plotly visualisasi data dan cara membuat plotly chart. 

```python
#memasukan modul yang digunakan
import plotly.offline as pyo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#membuka data
data = pd.read_excel('sales.xls')
data.head()
```

```python
#melihat info data
data.info()
```

```python
#membuat barchart dengan matplotlib
plt.plot(data.Sales)
plt.show()
```

```python
#membuat plotly chart 
#membuat barchart dengan plotly
pyo.plot([{
    'x' : data.index,
    'y' : data.Sales
}])

#hasilnya seperti gambar di bawah ini
```

```python
data1 = data[['Sales', 'Profit']]
data1.head()
```

```python
pyo.plot([{
    'x' : data1.index,
    'y' : data1[col],
    'name' : col} for col in data1.columns ])
```
