## Plotly - Automate Mutiple Line pada Line Chart di Plotly

Berikut ini cara meng-automate multiple line pada line chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#membuat pivot tabel dari kolom HomeGoals dan AwayGoals
fifa_hg = pd.pivot_table(fifa, values = ['HomeGoals', 'AwayGoals'], columns = 'Year')
fifa_hg.head()
```

```python
traces = [go.Scatter(
    x = fifa_hg.columns,
    y = fifa_hg.loc[rowname],
    mode = 'markers+lines',
    name = rowname
) for rowname in fifa_hg.index] 
```

```python
#mendefinisikan layout
layout = go.Layout(title = 'Average Away Goals and Home Goals by Year')
```

```python
#membuat frame
figure = go.Figure(data = traces, layout = layout)
```

```python
#melihat hasil
pyo.plot(figure, filename = 'Avg_Goals_Fifa_Traces.html')
```
