## Plotly - Penambahan Titik-Titik pada Line Chart di Plotly

Berikut ini cara menambahkan titik-titik pada line chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#membuat pivot tabel dari kolom HomeGoals dan AwayGoals
fifa_hg = pd.pivot_table(fifa, values = ['HomeGoals', 'AwayGoals'], index = 'Year')
fifa_hg.head()
```

```python
#mendefinisikan parameter scatter chart
trace0 = go.Scatter(
    x = fifa_hg.index,
    y = fifa_hg.AwayGoals, 
    mode = 'markers+lines',
    name = 'Away Goals'
)

trace1 = go.Scatter(
    x = fifa_hg.index,
    y = fifa_hg.HomeGoals, 
    mode = 'markers+lines',
    name = 'Home Goals'
)
```

```python
#mendefinisikan data di plotly
data = [trace0, trace1]
```

```python
#mendefinisikan layout
layout = go.Layout(title = 'Average Away Goals and Home Goals by Year')
```

```python
#membuat frame
figure = go.Figure(data = data, layout = layout)
```

```python
#melihat hasil
pyo.plot(figure, filename = 'Avg_Goals_Fifa_Markers_Lines.html')
```
