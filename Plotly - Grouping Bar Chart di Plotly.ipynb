## Plotly - Grouping Bar Chart di Plotly

Berikut ini cara membuat grouping bar chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#membuat pivot tabel dari kolom HomeGoals dan AwayGoals
fifa_hg = pd.pivot_table(fifa, values = ['HomeGoals', 'AwayGoals'], index = 'Year')
fifa_hg.head()
```

```python
#mendefinisikan parameter bar di plotly
trace0 = go.Bar(
    x = fifa_hg.index,
    y = fifa_hg.AwayGoals, 
)

trace1 = go.Bar(
    x = fifa_hg.index,
    y = fifa_hg.HomeGoals, 
)
```

```python
#mendefinisikan data di plotly
data = [trace0, trace1]
```

```python
#mendefinisikan layout
layout = go.Layout(title = 'Average Away Goals dan Home Goals by Year')
```

```python
#membuat frame
figure = go.Figure(data = data, layout = layout)
```

```python
#melihat hasil
pyo.plot(figure, filename = 'Avg_Goals_Fifa_Group_Bar.html')
```
