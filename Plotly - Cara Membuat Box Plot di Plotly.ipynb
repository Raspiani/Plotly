## Plotly - Cara Membuat Box Plot di Plotly
Berikut ini cara membuat box plot di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
orders = pd.read_excel('sales.xls')
orders.head()
```

```python
#melihat jumlah kategori di dalam variabel Customer Segment
orders['Customer Segment'].value_counts()
```

```python
#mendefiniskan masing-masing kategori
corp = orders.Sales[orders['Customer Segment'] == 'Corporate']
ho = orders.Sales[orders['Customer Segment'] == 'Home Office']
cons = orders.Sales[orders['Customer Segment'] == 'Consumer']
smb = orders.Sales[orders['Customer Segment'] == 'Small Business']
```

```python
#membuat trace
trace0 = go.Box(
    y = corp, 
    name = 'Corporate'
)

trace1 = go.Box(
    y = ho, 
    name = 'Home Office'
)

trace2 = go.Box(
    y = cons, 
    name = 'Consumer'
)

trace3 = go.Box(
    y = smb, 
    name = 'Small Business'
)
```

```python
#mendefinisikan data
data = [trace0, trace1, trace2, trace3]
```

```python
#membuat layout
layout = go.Layout(
    title = 'Distribution of Sales by Customer Segment',
)
```

```python
#mendefinisikan figure
figure = go.Figure(data = data, layout = layout)
```

```python
#melihat hasilnya
pyo.plot(figure, filename = 'Box Plot di Plotly.html')
```
