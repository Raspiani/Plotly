## Plotly - Membuat Sebuah Chart Bar Stack di Plotly
Berikut ini cara membuat sebuah chart bar stack di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#membuat pivot tabel dari kolom HomeGoals dan AwayGoals
fifa_hg = pd.pivot_table(fifa, values = ['HomeGoals', 'AwayGoals'], index = 'Year')
fifa_hg.head()
```

```python
#membuat trace
trace0 = go.Bar(
    x = fifa_hg.index,
    y = fifa_hg.AwayGoals
)

trace1 = go.Bar(
    x = fifa_hg.index,
    y = fifa_hg.HomeGoals
)
```

```python
#membuat data 
data = [trace0, trace1]
```

```python
#membuat layout
#tambahkan (..........., barmode='stack')
layout = go.Layout(title='Average Home Goals by Year', barmode='stack')
```

```python
#mmebuat figure 
figure = go.Figure(data = data, layout = layout)
```

```python
#menampilkan hasilnya
pyo.plot(figure, filename='Bar Stack di Plotly.html')
```
