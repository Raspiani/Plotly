## Plotly - Cara Customize Warna Chart di Plotly
Berikut ini cara customize warna chart di plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
#memasukan data
fifa = pd.read_excel('world_cup_results.xlsx')
fifa.head()
```

```python
#membuat pivot tabel dari kolom HomeGoals dan AwayGoals
fifa_hg = pd.pivot_table(fifa, values = ['HomeGoals', 'AwayGoals'], index = 'Year')
fifa_hg.head()
```

```python
#membuat trace
trace0 = go.Scatter(
    x = fifa_hg.index,
    y = fifa_hg.AwayGoals,
    mode = 'markers+lines',
    name = 'Away Goals'
)

#menambahkan marker = dict(color = 'rgb(190,190,190)')
#marker = dict(color = '#fa8072')

trace1 = go.Bar(
    x = fifa_hg.index,
    y = fifa_hg.HomeGoals,
    name = 'Home Goals', 
    marker = dict(color = '#fa8072')
)
```

```python
#membuat data 
data = [trace0, trace1]
```

```python
#membuat layout
layout = go.Layout(title='Average Home Goals by Year')
```

```python
#mmebuat figure 
figure = go.Figure(data = data, layout = layout)
```

```python
#menampilkan hasilnya
pyo.plot(figure, filename='Customize Bar dan Line Combo Chart di Plotly.html')
```
